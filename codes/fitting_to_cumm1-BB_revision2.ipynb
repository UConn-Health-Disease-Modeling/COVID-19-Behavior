```python
#Fixed parameter values
pi = 1710; 
#beta1 =1.72;
mu =1/(83*365); 
eps =0.33;   #Youngji paper
tau1 = 0.8; tau2 = 0.8; tau =0.8;
psi1 = 0.5; psi = 0.5;
phi1, phi2 = 1/180, 1/180
h1= 0.0045; 
h2 = (1-0.74)*h1;
rho = 1/14; rhot =1/14; rhoh =1/10;
eta1= 0.71; eta2= 0.41; eta4= 0.85; eta3 = 0.46;
sevr_im = 2;  # severity impact is higher than the prevalence impact, people react twice higher than the to prevalence info.
#beta1 = (1.72*(eps+mu)*(mu+rho)*(mu+rho+d1))/(eps*((tau1+tau2)*(mu+rho) +(1-tau1)*psi1*(mu+rho+d1)))


theta = 0.5 # parameter adjusting reduced perception risk (assumed)
delta = 0.8  # proportion of tested people who quarantine/adhere to protective measures (assumed)

#Dtilde, Btilde = 10, 20
D, B = 50, 50
# Fmax = 0.027; F10 = 0.0029; F20 = (0.1)*F10;
# #Tmax = 0.5; T10 = 0.03; T20 = 0.03
# Tmax = 0.1;  T10 = 0.0023; T20 = T10;  #0.022;
a = 1/3  # fixed to be 3 days 
a1 = 0.5
alpha1, alpha2 = 0.5, 0.5
```

```python
# Initial conditions
NN = 51815808
V2_0 = 2891 
V1_0 = 431 
E2_0 = 352645 #int(0.87*20*NewCases.iloc[0])
A2_0 = 70529 
I2_0 = 35264 
IT2_0 = 2026
E1_0 = 52694  # 0.13 indicates that 13% of the total population is in the first dynamics
A1_0 = 10538 
I1_0 = 5269 
IT1_0 = 2634 
IU1_0 = 0
IU2_0 = 0
H_0 = 202  # ICU
R_0 = 200
D_0 = 15

MV_0 = k1*((IT1_0 + IT2_0 + H_0)/NN)
MN_0 = k1*((H_0 + d1*(IU1_0 + IU2_0) + d2*H_0)/NN)
S2_0 = 40000000 
S3_0 = 4365186 
S1_0 = NN-(S2_0 + S3_0 + V1_0 + V2_0 + E1_0 + E2_0 + A1_0 + A2_0 + I1_0 + I2_0 + IT1_0 + IT2_0 + IU1_0 + IU2_0 + H_0 + R_0)

T1_0 = T10 + (Tmax -T10)*(alpha1*D*MV_0/(1+D*MV_0) + (1-alpha1)*B*MN_0/(1+B*MN_0) )
T2_0 = T20 + (Tmax -T20)*(alpha2*(theta*D*MV_0)/(1 + theta*D*MV_0) + (1-alpha2)*(theta*B*MN_0)/(1 + theta*B*MN_0) )
new_test_0 =  T1_0*(xi*A1_0 + I1_0) + T2_0*(xi*A2_0 + I2_0) 

NewDetected_0 = IT1_0 + IT2_0
NewVacc_0 = V1_0 + V2_0
```

```python
def model_eq(x,tf, beta1, beta2, k1, xi):
    S1, S2, S3, V1, V2, E1, E2, A1, A2, I1, I2, I_T1, I_T2, I_U1, I_U2, H, R, Dth, MV, MN, Newdetected, Newvacc = x
    #dx = np.zeros(9)
    
    
    N= S1 + S2+ S3 + V1 + V2 + E1 + E2 + A1 + A2 + I1 + I2 +  I_T1 + I_T2 + I_U1 + I_U2 + H + R
    
    T1 = T10 + (Tmax -T10)*(alpha1*(D*MV)/(1+D*MV) + (1-alpha1)*(sevr_im*B*MN)/(1+sevr_im*B*MN) )
    F1 = F10 + (Fmax -F10)*(alpha1*(D*MV)/(1+D*MV) + (1-alpha1)*(sevr_im*B*MN)/(1+sevr_im*B*MN))
    
    T2 = T20 + (Tmax -T20)*(alpha2*(theta*D*MV)/(1+theta*D*MV) + (1-alpha2)*(theta*sevr_im*B*MN)/(1+theta*sevr_im*B*MN) )
    F2 = F20 + (Fmax -F20)*(alpha2*(theta*D*MV)/(1+theta*D*MV) + (1-alpha2)*(theta*sevr_im*B*MN)/(1+theta*sevr_im*B*MN))
    
    L1 = beta1*(I_U1 + I_U2 + (1-delta)*(I_T1 + I_T2))/(N-(delta*(I_T1 + I_T2) + H))
    L2 = beta2*(I_U1 + I_U2 +(1-delta)*(I_T1 + I_T2))/(N-(delta*(I_T1 + I_T2) + H))
    
    dS1 = pi-(F1 +L1 +mu)*S1
    dS2 = phi1*V1 -((1-eta2)*L2+F2+mu)*S2
    dS3 = phi2*R-((1-eta3)*L2+F2+mu)*S3
    dV1 = F1*S1-phi1*V1-(1-eta1)*L2*V1-mu*V1
    dV2 = F2*S2+F2*S3-(1-eta4)*L2*V2-mu*V2
    dE1 = L1*S1-(eps+mu)*E1
    dE2 = (1-eta1)*L2*V1+(1-eta4)*L2*V2+L2*((1-eta2)*S2+(1-eta3)*S3)-(mu+eps)*E2
    dA1 = (1-tau1)*eps*E1 - a1*A1 
    dA2 = (1-tau2)*eps*E2 - a1*A2 
    dI1 = tau1*eps*E1 - I1 
    dI2 = tau2*eps*E2 - I2 
    dI_T1 = a1*xi*T1*A1+ T1*I1 -(rho + h1 + mu)*I_T1
    dI_T2 = a1*xi*T2*A2 + T2*I2-(rho + h2 + mu)*I_T2
    dI_U1 = a1*(1-xi*T1)*A1 + (1-T1)*I1 -(rho + mu)*I_U1 #13
    dI_U2 = a1*(1-xi*T2)*A2+ (1-T2)*I2 - (rho + mu)*I_U2 #14
    dH = h1*I_T1 + h2*I_T2 - (rhoh + d2 + mu)*H
    dR = rho*(I_U1 + I_U2 + I_T1 + I_T2) + rhoh*H -(phi2 + mu)*R
    dDth = d1*(I_U1+I_U2) + d2*H  #17
    dMV = a*(k1*((I_T1 + I_T2 + H)/NN) - MV)  #18
    dMN = a*(k1*((H + d1*(I_U1 + I_U2) + d2*H)/NN)-MN)  #19
    #dMN = a*(k1*((H + Dth)/NN)-MN) 
    dDetected = xi*T1*A1 + T1*I1 + xi*T2*A2 + T2*I2
    dNewVacc = F1*S1 + F2*S2 + F2*S3
    return dS1, dS2, dS3, dV1, dV2, dE1, dE2, dA1, dA2, dI1, dI2, dI_T1, dI_T2, dI_U1, dI_U2, dH, dR, dDth, dMV, dMN, dDetected, dNewVacc
```

```python
def solve_ode(tf, beta1, beta2, k1, xi):
    y = odeint(model_eq, x0, tf,args = (beta1, beta2, k1, xi))   
    
    Vac1_1 = F10 + (Fmax -F10)*(alpha1* np.divide(D*y[:,18],(1+D*y[:,18])) + (1-alpha1)* np.divide(sevr_im*B*y[:,19],(1+sevr_im*B*y[:,19])))  
    Vac2_1 = F20 + (Fmax -F20)*(alpha2* np.divide(theta*D*y[:,18],(1+theta*D*y[:,18])) + (1-alpha2)* np.divide(theta*sevr_im*B*y[:,19],(1+theta*sevr_im*B*y[:,19])) ) 
    new_vac = np.multiply(Vac1_1,y[:,0]) + np.multiply(Vac2_1,(y[:,1]+y[:,2])) # 

    Test1_1 = T10 + (Tmax -T10)*(alpha1* np.divide(D*y[:,18],(1+D*y[:,18])) + (1-alpha1)* np.divide(sevr_im*B*y[:,19],(1+sevr_im*B*y[:,19]))) 
    Test2_1 = T20 + (Tmax -T20)*(alpha2* np.divide(theta*D*y[:,18],(1+theta*D*y[:,18])) + (1-alpha2)* np.divide(theta*sevr_im*B*y[:,19],(1+theta*sevr_im*B*y[:,19]))) 
    
    new_test =  np.multiply(Test1_1,(a1*xi*y[:,7]+ y[:,9])) + np.multiply(Test2_1,(a1*xi*y[:,8]+ y[:,10]))#
    #Daily_tested = np.diff(y[:,11]+y[:,12])
    
    new_death1 = d1*(y[:,13]+y[:,14]) + d2*y[:,15] 

    # NNN = y[:,0] + y[:,1]  + y[:,2] + y[:,3] + y[:,4] + y[:,5] +y[:,6] + y[:,7] + y[:,8]  + y[:,9] + y[:,10] + y[:,14]
    # L11 = np.divide(beta1*(y[:,9] + y[:,10] + psi1*(y[:,7] + y[:,8])),NNN)
    # L22 = np.divide(beta2*(y[:,9] + y[:,10] + psi1*(y[:,7] + y[:,8])),NNN)
    # new_inf = np.multiply(L11,y[:,0]) + np.multiply(L22,(y[:,1] + y[:,2])) + (1-eta1)*np.multiply(L11,y[:,3]) + (1-eta2)*np.multiply(L22,y[:,4])  


    #App_cum_cases = np.cumsum(y[:,9] + y[:,10])
    #App_cum_vac = np.cumsum(y[:,3] + y[:,4])
    App_cum_cases = np.cumsum(new_test)
    App_cum_vac = np.cumsum(new_vac)
    #App_cum_test = np.cumsum(new_test)
    App_cum_death = np.cumsum(new_death1)
    return App_cum_cases, App_cum_vac, App_cum_death
```

```python
def fit_model(tf,beta1, beta2, k1, xi):
    apr_cum_cases, apr_cum_vac, apr_cum_death = solve_ode(tf, beta1, beta2, k1, xi)
    return np.concatenate([apr_cum_cases, apr_cum_vac, apr_cum_death])
    
```

```python
def sum_squared_error(params, tf, cum_new_cases, cum_new_vacc, cum_new_death):
    beta1, beta2, k1, xi = params
    y_fit_infected, y_fit_vacc, y_fit_death = solve_ode(tf, beta1, beta2, k1, xi)
    error_inf = cum_new_cases - y_fit_infected
    error_vacc = cum_new_vacc - y_fit_vacc
    error_death =  cum_new_death - y_fit_death
    return np.sum(error_inf**2) + np.sum(error_vacc**2) + np.sum(error_death**2)
```

```python
official_cum_data = np.concatenate([cum_new_cases,cum_new_vacc,cum_new_death])  
```

```python
tolerance = 1
iteration = 0
max_iterations = 1
current_params = [beta1, beta2, k1, xi]

while iteration < max_iterations:
    # Use curve_fit to fit the SIR model to the data.
    #  [(0,   0,   0,   0.5,  0,   0,   1/5,  0,     0.02,     1.8e-5,   1.8e-5,     0.8,      5e-6,  0.5),\
    # (2,   100, 100, 0.9,  100, 100, 1/2,  0.9,   0.27,     1.9e-2,   1.9e-2,     0.9,    0.75,  0.75)]
    
    popt, pcov = curve_fit(fit_model,tf,official_cum_data,p0 = current_params,bounds =[(0, 0, 0.1, 0.1),
                                                                                       (10, 10, 0.9, 1)])   
    current_error = sum_squared_error(popt, tf, cum_new_cases,cum_new_vacc,cum_new_death) #cum_new_cases

    # Check if the sum squared error is below the tolerance.
    if current_error < tolerance:
        break

    # Update the initial parameters with the current parameters.
    current_params = popt

    iteration += 1
print('Maximum iteration =', iteration)  
#print('Current error = ',current_error)
fitted_cum_cases, fitted_cum_vacc, fitted_cum_death = solve_ode(tf, *popt)

beta1 = popt[0]; beta2 = popt[1];  k1 = popt[2];  xi = popt[3]; #d1 = popt[4];  d2 = popt[5];
# Fmax = popt[4]; F10 = popt[5];  F20 = popt[6];  Tmax = popt[7];
# T10 = popt[8]; T20 = popt[9];   


S1_dfe = pi/(F10 +mu); S2_dfe = (pi*phi1*F10)/((phi1+mu)*(F10+mu)*(F20+mu)); V1_dfe = (pi*F10)/((phi1+mu)*(F10 +mu));
V2_dfe = (pi*phi1*F10*F20)/(mu*(phi1+mu)*(F10+mu)*(F20+mu));
N_dfe = S1_dfe + S2_dfe + V1_dfe + V2_dfe;

C1 = (beta1*S1_dfe)/N_dfe;  C2 = (1/N_dfe)*(beta2*((1-eta2)*S2_dfe + (1-eta4)*V2_dfe) + (1-eta1)*V1_dfe);

A11 = (1-delta)*T10*(eps/(eps + mu))*(((1-tau)*xi + tau)/(h1 + mu + rho))
A12 = (eps/(eps + mu))*(((xi-1)*tau*T10 + (1-T10*xi))/(mu + rho + d1))


A21 = (1-delta)*T20*(eps/(eps + mu))*(((1-tau)*xi + tau)/(h2 + mu + rho))
A22 = (eps/(eps + mu))*(((xi-1)*tau*T20 + (1-T20*xi))/(mu + rho + d1))

R_e = C1*(A11 + A12) + C2*(A21 + A22)
print('Reproduction number =', R_e)
```

```python
# Estimated parameters
beta1, beta2, k1, xi = popt

# Standard errors from covariance matrix
perr = np.sqrt(np.diag(pcov))

# 95% confidence intervals
ci_lower = popt - 1.96 * perr
ci_upper = popt + 1.96 * perr

print("\nParameter estimates with 95% CI:")
param_names = ["beta1", "beta2", "k1", "xi"]
for i, name in enumerate(param_names):
    print(f"{name}: {popt[i]:.4f}  (95% CI: {ci_lower[i]:.4f}, {ci_upper[i]:.4f})")
```

```python
rmse_cases = np.sqrt(np.mean((cum_new_cases - fitted_cum_cases)**2))
rmse_vacc  = np.sqrt(np.mean((cum_new_vacc - fitted_cum_vacc)**2))
rmse_death = np.sqrt(np.mean((cum_new_death - fitted_cum_death)**2))

print("RMSE Cases:", rmse_cases)
print("RMSE Vaccinations:", rmse_vacc)
print("RMSE Deaths:", rmse_death)
```

```python
# 1. Minmax normalization (0-1 scale)
def normalized_rmse_minmax(obs, pred):
    obs_norm = (obs - np.min(obs)) / (np.max(obs) - np.min(obs))
    pred_norm = (pred - np.min(obs)) / (np.max(obs) - np.min(obs))
    return np.sqrt(np.mean((obs_norm - pred_norm)**2))

nrmse_cases = normalized_rmse_minmax(cum_new_cases, fitted_cum_cases)
nrmse_vacc  = normalized_rmse_minmax(cum_new_vacc, fitted_cum_vacc)
nrmse_death = normalized_rmse_minmax(cum_new_death, fitted_cum_death)

print("Normalized RMSE (min-max) Cases:", nrmse_cases)
print("Normalized RMSE (min-max) Vaccinations:", nrmse_vacc)
print("Normalized RMSE (min-max) Deaths:", nrmse_death)

# 2. Relative RMSE (RMSE divided by mean observed)
rrmse_cases = rmse_cases / np.mean(cum_new_cases)
rrmse_vacc  = rmse_vacc / np.mean(cum_new_vacc)
rrmse_death = rmse_death / np.mean(cum_new_death)

print("Relative RMSE Cases:", rrmse_cases)
print("Relative RMSE Vaccinations:", rrmse_vacc)
print("Relative RMSE Deaths:", rrmse_death)
```
